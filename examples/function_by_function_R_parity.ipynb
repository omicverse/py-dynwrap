```python
import sys; sys.path.insert(0, '..')
import pydynwrap as dw
import pandas as pd, numpy as np

# wrap_data
ds = dw.wrap_data(['a','b','c'])
print('R: wrap_data(c("a","b","c"))  →  cell_ids:', ds.cell_ids)

# add_linear_trajectory
ds = dw.wrap_data([f'c{i}' for i in range(6)])
ds = dw.add_linear_trajectory(ds, np.linspace(0,1,6))
print('R: add_linear_trajectory(ds, pseudotime=seq(0,1,length.out=6))')
print('   milestones:', ds.milestone_ids)
print('   pseudotime:', ds.pseudotime.values)
```
