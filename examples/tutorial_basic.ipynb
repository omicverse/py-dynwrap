# pydynwrap tutorial — 3-branch trajectory

Walks through every public function on a deterministic synthetic.

```python
import sys
sys.path.insert(0, '..')
import pydynwrap as dw
import numpy as np, pandas as pd
print(f'pydynwrap {dw.__version__}')
```

## wrap_data → add_branch_trajectory → calculate_geodesic_distances

```python
ds = dw.wrap_data(cell_ids=[f'c{i}' for i in range(30)])
ds = dw.add_branch_trajectory(ds,
    pseudotime=np.tile(np.linspace(0,1,10), 3),
    branch=np.repeat(['A','B','C'], 10))
print('milestone_ids:', ds.milestone_ids)
print('progressions head:'); print(ds.progressions.head())
print('milestone_percentages head:'); print(ds.milestone_percentages.head())
```

## classify_milestone_network

```python
print(dw.classify_milestone_network(ds.milestone_network))
```

## calculate_geodesic_distances

```python
gd = dw.calculate_geodesic_distances(ds, directed=False)
print('gd shape:', gd.shape); gd.iloc[:5, :5]
```

## add_dimred / add_pseudotime / add_grouping / add_expression

```python
rng = np.random.RandomState(0)
ds = dw.add_dimred(ds, rng.randn(30, 2))
ds = dw.add_expression(ds, rng.poisson(2, (30, 100)))
print('has dimred:', dw.is_wrapper_with_dimred(ds))
print('has expression:', dw.is_wrapper_with_expression(ds))
```
